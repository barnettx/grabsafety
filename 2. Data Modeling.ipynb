```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math

from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import SGDClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report, confusion_matrix

import seaborn as sns
from sklearn.externals import joblib
```

```python
data_combined = pd.read_csv('./data_combined.csv', header=0, index_col=0)
```

```python
data_combined.sort_values(["bookingID","second"], inplace=True)
```

## Read Data

```python
X = data_combined[["bookingID",
      "accuracy",
      "bearing",
      "acceleration_x","acceleration_y","acceleration_z",
      "gyro_x","gryo_y","gyro_z",
      "second",
      "speed"]]
y = data_combined[["label"]]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

## Standardise data

```python
scaler = StandardScaler().fit(X_train)
standardized_X = scaler.transform(X_train)
standardized_X_test = scaler.transform(X_test)
```

Save our scaler for predictions

```python
joblib.dump(scaler, 'standard_scaler.pkl')
```

# Build Models
Referencing: https://scikit-learn.org/stable/tutorial/machine_learning_map/index.html
Testing with Stochastic Descent Classifier

```python
sgd = SGDClassifier(verbose=100, alpha=1e-4, loss='hinge', max_iter=1000)
history = sgd.fit(standardized_X, y_train)
```

```python
sgd.score(standardized_X_test, y_test)
```

```python
y_pred = sgd.predict(standardized_X_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
joblib.dump(sgd, 'sgd_1000_all_data.pkl')
```
